```python
def keep_rows_year(row, year):
    if row['YearOut']:
        return (year <= row['YearOut']) & (year >= row['YearIn'])
    else:
        return (year >= row['YearIn'])
# c’est le boléan créé pour détecter si les gens qui sont présentes dans l’année désignée.
list_all_groupePolitique = graphique_avg['GroupNameFR'].unique()
list_all_groupePolitique
# c’est pour avoir une liste des groupe politique
```

```python
dic_ageMoyen_parGroupePolitique = {}
for year in range(1949, 2025):
    print(year)
    df_year = graphique_avg[graphique_avg.apply(keep_rows_year, args=(year,), axis=1)]

    for groupPolitiqueName in list_all_groupePolitique:
        df_groupepolitique = df_year[df_year['GroupNameFR'] == groupPolitiqueName]
        df_homme_ageM = df_groupepolitique[df_groupepolitique['SexFR'] == 'Homme']
        ageMoyen_hommes = df_homme_ageM['age_in_position'].mean()
        df_femme_ageM = df_groupepolitique[df_groupepolitique['SexFR'] == 'Femme']
        ageMoyen_femmes = df_femme_ageM['age_in_position'].mean()
        dic_ageMoyen_parGroupePolitique[(year, groupPolitiqueName)] = (ageMoyen_hommes, ageMoyen_femmes)
dic_ageMoyen_parGroupeP = pd.DataFrame(dic_ageMoyen_parGroupePolitique)
```
